## 2. Variables and Basic Data Types

A **variable** is like a labeled box where you can store information. You give the box a name (the variable name) and put something inside it (the value).

```python
# In Python, you create a variable by assigning a value to it with the '=' sign.

student_name = "Alex"
age = 21
pi_value = 3.14
is_learning = True

# We can print the values stored in these variables
print(student_name)
print(age)
print(pi_value)
print(is_learning)
```

### Common Data Types
Python has several built-in data types. Here are the most common ones:

- **String (`str`):** Text, enclosed in single `' '` or double `" "` quotes. (e.g., `"Hello World"`)
- **Integer (`int`):** Whole numbers, positive or negative. (e.g., `10`, `-50`)
- **Float (`float`):** Numbers with a decimal point. (e.g., `3.14`, `-0.001`)
- **Boolean (`bool`):** Represents one of two values: `True` or `False`.

You can check the type of any variable using the `type()` function.

```python
print(type("This is a string"))
print(type(100))
print(type(99.9))
print(type(False))
```

## 3. Arithmetic Operations

Python can be used as a powerful calculator.

```python
a = 10
b = 3

# Addition
print(f"{a} + {b} = {a + b}")

# Subtraction
print(f"{a} - {b} = {a - b}")

# Multiplication
print(f"{a} * {b} = {a * b}")

# Division
print(f"{a} / {b} = {a / b}")

# Floor Division (discards the remainder)
print(f"{a} // {b} = {a // b}")

# Modulo (returns the remainder)
print(f"{a} % {b} = {a % b}")

# Exponent (to the power of)
print(f"{a} ** {b} = {a ** b}")
```

## 4. Input and Output

A program becomes much more useful when it can interact with a user.

### Output with `print()`

We've already used `print()` to display variable values. It can also print text and combine text with variables, especially using f-strings (`f"..."`).

```python
name = "Artur"
age = 35

# Using an f-string to format the output
print(f"My name is {name} and I am {age} years old.")
```

### Input with `input()`

The `input()` function pauses your program and waits for the user to type something and press Enter.

**Crucial Point:** `input()` **always** returns the user's input as a **string**, even if they type numbers!

```python
user_name = input("Please enter your name: ")
print(f"Hello, {user_name}!")
```

### Converting Input for Calculations

Because `input()` gives us a string, we need to convert it to a number (`int` or `float`) before we can use it in math operations. We do this with the `int()` and `float()` functions.

```python
age_str = input("How old are you? ")

# Convert the string to an integer
age_int = int(age_str)

age_in_10_years = age_int + 10
print(f"In 10 years, you will be {age_in_10_years} years old.")
```

### Simple Area Calculator

Let's write a program that calculates the area of a rectangle.
1.  Ask the user to enter the width of the rectangle.
2.  Ask the user to enter the height of the rectangle.
3.  Convert both inputs to numbers (you can use `float` to allow for decimals).
4.  Calculate the area (Area = width * height).
5.  Print the result in a user-friendly message.

```python
# Your code here

# 1. & 2. Ask for width and height
width_str = input("Enter the width of the rectangle: ")
height_str = input("Enter the height of the rectangle: ")

# 3. Convert inputs to floats
width = float(width_str)
height = float(height_str)

# 4. Calculate the area
area = width * height

# 5. Print the result
print(f"The area of a rectangle with width {width} and height {height} is {area}.")
```
